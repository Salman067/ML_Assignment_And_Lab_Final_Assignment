```python
# Import essential libraries
from numpy import loadtxt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
import seaborn as sns
```

```python
#Load data
data_path = '/content/drive/MyDrive/Normalaization/Datasets/House_Rent_Dataset.csv'
dataset = pd.read_csv(data_path)
dataset.shape
print(dataset.head())
```

```python
# split into input (X) and output (y) variables
X = dataset[['BHK', 'Size', 'Bathroom']]
y = dataset['Rent']

X.tail()
```

```python
#Normalization of input feature
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)



#Preparing train and test data
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)


#Define the neural network model
model = Sequential()
model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python

#Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

#Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Evaluate the model on the test set
mse = model.evaluate(X_test, y_test)
print(f'Mean Squared Error on Test Set: {mse}')
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)
```

```python

# Display some predictions and actual values
results = pd.DataFrame({'Actual Rent': y_test, 'Predicted Rent': predictions.flatten()})
print(results.tail())
```
