```python
data_path = '/content/drive/MyDrive/Normalaization/Datasets/diabetes.csv'
dataset = pd.read_csv(data_path, header=None)
dataset.shape
print(dataset.tail())
```

```python
from sklearn.preprocessing import StandardScaler

X = dataset.iloc[:, 0:8]
y = dataset.iloc[:, 8]

X.tail()
```

```python
# Ensure that X contains only numeric values
X_numeric = X.apply(pd.to_numeric, errors='coerce').fillna(0)

# Normalize the input features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X_numeric)

X_normalized
```

```python
X = X.apply(pd.to_numeric, errors='coerce').fillna(0)
y = y.apply(pd.to_numeric, errors='coerce').fillna(0)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(8, input_shape=(8,), activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=100, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# fit the keras model on the dataset for normalized data
model.fit(X_normalized, y, epochs=100, batch_size=10)
```

```python
# evaluate the keras model for normalized data
_, accuracy = model.evaluate(X_normalized, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)
# summarize the first 5 cases
for i in range(5):
	print(' %d (expected %d)' % ( predictions[i], y[i]))
```
